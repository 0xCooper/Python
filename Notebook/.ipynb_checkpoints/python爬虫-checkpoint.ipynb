#### 简单的第一个爬虫


```python
# 获取页面
import requests
link ="http://101.200.46.43/web/"  # 目标网页地址
headers ={'User-Agent': 'Mozilla/5.0(Windows;U;Windows NT 6.1;en-US; rv:1.9.1.6)Gecko/20091201 Firefox/3.55.6'}#定义请求头的浏览器代理，伪装成浏览器
r = requests.get(link,headers =headers)
print(r.text)#获取网页内容代码
```
