#### numpy的基础运算

```python
import numpy as np
a = np.array([1.57,20,30,40])
b = np.arange(4)
print ("a",a)
print ("b",b)
c = a-b
print(c)
c=c**2#c的平方
print("c",c)



c = 10*np.sin(a)#a的正弦值加+a
print(c)
print("b<3",b<3)#打印出b<3的bool值
print(b==3)#打印出b等于3的bool值
```

```python
import numpy as np
a = np.array([[0,1],[0,1]])
print(a)
b = np.arange(4).reshape((2,2))
print(b)

c=a*b#逐个相乘
c_dot =np.dot(a,b)#矩阵的运算
c_dot_2=a.dot(b)
print(c)
print(c_dot)
print(c_dot_2)
```

```python

import numpy as np
a= np.random.random((2,4))#随机数字
print(a)
np.sum(a,axis=1)#求和  axis维度为1 在列中求和 为0的话为行中qiuhe

np.min(a)#最小值
np.max(a)#最大值
```

#### matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(1,20,50)
y=x**2
plt.figure(num=1)
plt.xlim(0,50)

plt.yticks([0,100,250,300,500],["terrible","bad","normal","better","height"])
plt.plot(x,y)


plt.figure(num=2)
plt.plot(x,color='red')
plt.show()


```
